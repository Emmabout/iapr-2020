### 1.2 Region growing
We implement region growing by first thresholding the image (this replaces the homogeneity criterion), and then growing the region starting from an origin point manually selected.

First, threshold the image:

```python
# Manually chosen point of reference in the brain region
orig = [157, 80]
# reference intensity around this point
base_lev = np.mean(brain_im[152:162,75:85])
thresh = 20
min_thresh = base_lev - 30
max_thresh = base_lev + 20

# threshold image around reference intensity
mask = np.logical_and(brain_im >= base_lev - thresh, brain_im <= base_lev + thresh)
```

Display the thresholded image:
    

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.imshow(brain_im, cmap='gray')
ax.imshow(mask)
ax.set_title('MRI brain image ({} px, {} px)'.format(im_h, im_w))
ax.axis('off')
plt.plot([orig[1]],[orig[0]], marker="x")
plt.show()
```

Now, grow region starting from reference point:


```python
to_be_explored = [tuple(orig)]
explored = set()
brain = set()
while len(to_be_explored):
    i,j = to_be_explored.pop()
    explored.add((i,j))
    if not (min_thresh <= brain_im[i,j] <= max_thresh):
        continue
    brain.add((i,j))
    if i > 0:
        if (i-1, j) not in explored:
            to_be_explored.append((i-1, j))
    if j > 0:
        if (i, j-1) not in explored:
            to_be_explored.append((i, j-1))
    if i < 255:
        if (i+1, j) not in explored:
            to_be_explored.append((i+1, j))
    if j < 255:
        if (i, j+1) not in explored:
            to_be_explored.append((i, j+1))

print("Nb of pixels in brain:", len(brain)) 


brain_mask = np.array(([[(i,j) in brain for j in range(256)] for i in range(256)]))
```

Display the grown brain region:

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# display original image
axs[0].imshow(brain_im, cmap='gray')
axs[0].set_title('MRI brain image ({} px, {} px)'.format(im_h, im_w))
axs[0].axis('off')

# display grown brain region
axs[1].imshow(brain_mask)
axs[1].set_title('Grown brain region')
axs[1].axis('off')

plt.show()
```

As we can see, the brain is almost perfect but for a small region between the nose and eye. Let's look at this region in more detail.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# display original image
axs[0].imshow(brain_im, cmap='gray')
axs[0].set_title('MRI brain image - zoom')
axs[0].plot([158], [83], marker='x')
axs[0].axis('off')
axs[0].set_xlim((140, 170))
axs[0].set_ylim((100, 50))

# display grown brain region
axs[1].imshow(brain_mask)
axs[1].set_title('Grown brain region - zoom')
axs[1].plot([158], [83], marker='x')
axs[1].axis('off')
axs[1].set_xlim((140, 170))
axs[1].set_ylim((100, 50))

plt.show()
```

Apparently this extra region added to the brain is due to one single pixel that connects it to the brain.

This problem might possibly be removed by tuning the threshold, but this would have to be done by hand and would not be robust.

However, knowing the structure of the brain, it is reasonable to suppose that the brain is only made of (relatively) thick parts. Therefore we can decide that any pixel that is the only connection between two regions of the brain should be removed, thus disconnecting the regions.

To implement this, we use the function skimage.measure.label, which labels each pixel with a region id (one region id per connected region of equal values).
We try removing each pixel of the brain, and count the connected regions resulting from this removal. If one additional region is created by removing the pixel, then the pixel is permanently removed from the brain.
Finally we keep only the main brain region (the one that contains the reference point).

```python
# function that counts the number of connected regions:
def nb_labels(mask):
    ls = set()
    labs = skimage.measure.label(mask)
    for l in (j for line in labs for j in line):
        ls.add(l)
    return len(ls)
```

```python
# copy the brain mask
brain_mask_connected = np.array([[v for v in line] for line in brain_mask])
nb_labels(brain_mask_connected)
```

So far their are only two regions (brain and non-brain).

Now we remove from brain_mask_connected the pixels that connect two distinct regions.

```python
for (i,j) in brain:
    save_val = brain_mask[i,j]
    # try removing pixel (i,j)
    brain_mask[i,j] = False
    # count resulting regions
    n_reg = nb_labels(brain_mask)
    if n_reg > 2:   # in this case the pixel removal increased the number of regions
        # remove this pixel
        brain_mask_connected[i,j] = False
    else:
        # keep this pixel
        brain_mask_connected[i,j] = save_val
    # restore original brain_mask
    brain_mask[i,j] = save_val
```

Now, label the resulting connected regions and only retain the one that contains the reference pixel.

```python
# label resulting regions
labs = skimage.measure.label(brain_mask_connected)

# keep only the right region
good_lab = labs[tuple(orig)]
good_brain_mask = labs==good_lab
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# display original image
axs[0].imshow(brain_im, cmap='gray')
axs[0].set_title('Original MRI brain image')
axs[0].axis('off')

# display grown brain region
axs[1].imshow(good_brain_mask)
axs[1].set_title('Selected brain region')
axs[1].axis('off')

plt.show()
```

Finally, compute the size of the brain:

```python
print("Number of pixels in brain:", np.sum(good_brain_mask))
```

### 1.3 Contour detection

The function `edge_detection` takes as parameters :
- `image` (the image that we want to analyse)
- `index` is the index of the labelled region that we want. Here, the brain. The index is found manually.
- `Sobel` is a boolean : 1 means that the Sobel method is used, 0 is Prewitt.
- `threshold` is the threshold for the mask.

Four matrices are used (`h1`, `h2`, `h3`, `h4`), in order to find horizontal and vertical edges. We thought the results were good with those four matrices. 

```python
def edge_detection(image, index, Sobel, threshold):
    im_h, im_w = image.shape
    
    if Sobel: #use the Sobel matrices
        h1 = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
        h2 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
        h3 = -h1
        h4 = -h2
        
    else: #Prewitt
        h1 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
        h2 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
        h3 = -h1
        h4 = -h2
    
    edge1 = np.zeros((im_h,im_w))
    edge2 = np.zeros((im_h,im_w))
    edge3 = np.zeros((im_h,im_w))
    edge4 = np.zeros((im_h,im_w))
    
    for i in range (1,im_h - 1):
        for j in range (1,im_w - 1):
            
            #create a small image containing 9 pixels, to multiply with h1,h2,h3,h4
            small_img = np.array([[image[i-1][j-1], image[i-1][j], image[i-1][j+1]],
                                  [image[i][j-1], image[i][j], image[i][j+1]],
                                  [image[i+1][j-1], image[i+1][j], image[i+1][j+1]]])
            #multiply the small image with h1,h2,h3,h4 and sum the result for each pixel and for each matrix
            edge1[i][j] = np.sum(np.multiply(small_img,h1))
            edge2[i][j] = np.sum(np.multiply(small_img,h2))
            edge3[i][j] = np.sum(np.multiply(small_img,h3))
            edge4[i][j] = np.sum(np.multiply(small_img,h4))
            
    edge1 = np.abs(edge1)
    edge2 = np.abs(edge2)
    edge3 = np.abs(edge3)
    edge4 = np.abs(edge4)
    
    edge = (edge1 + edge2 + edge3 + edge4)/4
    #We want to keep the edges that are under a certain threshold (tuned manually)
    mask = edge[:,:] < threshold
    
    #label labels the different regions of the image, defined by the edges.
    labelled = label(mask)
    
    edge[mask] = 0
    edge[np.logical_not(mask)] = 255
    
    fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(9, 5),
                         sharex=True, sharey=True)
    ax = axes.ravel()
    ax[0].imshow(image, cmap='gray')
    ax[0].set_title('Image')
    
    ax[1].imshow(edge, cmap='gray')
    ax[1].set_title('edges')
    
    #plot the different labelled regions
    ax[2].imshow(labelled, cmap='hsv')
    ax[2].plot([80], [157], marker='x')
    ax[2].set_title('Labels')
    
    
    for a in ax:
        a.axis('off')
    fig.tight_layout()
    
    #manually, we found the label corresponding to the brain. We set the value of those pixels to 255.
    brain = np.zeros((im_h,im_w))
    mask = labelled[:,:] == index
    brain[mask] = 255
    brain[np.logical_not(mask)] = 0    
    
    ax[3].imshow(brain, cmap='gray')
    ax[3].set_title('Brain')
    
    plt.show()
    
    #pixels are either 0 (no brain) or 255 (brain)
    #in order to obtain the number of pixels of brain, we just divide the sum of `brain` by 255.
    return np.sum(brain)/255
```

## Prewitt method

```python
index = 153
brain_size_prewitt = edge_detection(brain_im, index,0, 30)
print("Brain size using Prewitt (pixels) : ", brain_size_prewitt)
```

## Sobel method

```python
index = 138
brain_size_sobel = edge_detection(brain_im, index,1,41)
print("Brain size using sobel (pixels) : ", brain_size_sobel)
```

# Intensity normalization
As the three images have different luminosities, it is easier for us to normalize them, in order to be able to use the same methods on all images.

We used the method of Contrast streching, with the function `exposure_rescale_intensity`, tuning the percentiles by trial an error. 

The result is three images that look more alike than the original images. The normalized images are stored in an array `new_images`.

```python
# Contrast stretching : percentiles
p2, p98 = np.percentile(images[0], (1, 98))
p21, p981 = np.percentile(images[1], (2, 98))
p22, p982 = np.percentile(images[2], (1, 98))

#rescale intensity
newim1 = exposure.rescale_intensity(images[0], in_range=(p2, p98))
newim2 = exposure.rescale_intensity(images[1], in_range=(p21, p981))
newim3 = exposure.rescale_intensity(images[2], in_range=(p22, p982))

#contains the new images. dimensions[3 480 640 3]
new_images = np.array([newim1, newim2, newim3])

# Display images
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(18,10),
                     sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(newim1, cmap='gray')
ax[0].set_title('arena-shapes-01, normalized')

ax[1].imshow(newim2, cmap='gray')
ax[1].set_title('arena-shapes-02, normalized')

ax[2].imshow(newim3, cmap='gray')
ax[2].set_title('arena-shapes-03, normalized')

for a in ax:
    a.axis('off')

plt.show()
```

`get_shapes` is a function that gets the contours of the shapes, from the normalized images. The contours of the shapes are obtained thanks to `measure.find_contours` : this functions finds contours in an image, given a level. After trial and error, we found that a level of 100 would give good results. A smaller level does not work, because the function finds shapes within the shapes (due to the noise in the images). 

After finding the contours, `get_shapes` also finds the center of each contour, which is the center of each shape. This will be useful later to find the color of each shape. 

The function displays the results (contours and centers of shapes) for the three images.

```python
def get_shapes(image):
    contours = measure.find_contours(image, 100)

    labels=measure.label(image,connectivity=2)
    dst=color.label2rgb(labels)

    fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(16, 24))
    ax0.imshow(image, plt.cm.gray)
    ax1.imshow(image, plt.cm.gray)
    
    #list with coordinates of the center of each shape
    list_centers = np.zeros((len(contours),2))
    
    #The centers of the shapes are determined by calculating the mean of all coordinates of the contour.
    for n, contour in enumerate(contours):
        ax1.plot(contour[:, 1], contour[:, 0], linewidth=2)
        list_centers[n][0] = np.mean(contours[n][:,0])
        list_centers[n][1] = np.mean(contours[n][:,1])

    contours = np.array(contours)
    
    for i in range (0,len(list_centers)):
        ax1.plot([np.mean(contours[i][:,1])], [np.mean(contours[i][:,0])], marker='x')
    
    plt.show() 
    
    #need to round list_centers to have coordinates of pixels.
    return np.round(list_centers), contours
```

```python
list_centers1, contours1 = get_shapes(new_images[0,:,:,0])
list_centers2, contours2 = get_shapes(new_images[1,:,:,0])
list_centers3, contours3 = get_shapes(new_images[2,:,:,0])

contours = [contours1, contours2, contours3]

for i in range(0,len(contours)):
    print("arena-shapes-0%i"%(i+1), ":", len(contours[i]), "shapes detected")

list_centers = np.array([list_centers1, list_centers2, list_centers3])
```

`draw_contours` is drawing the contours and filling them in a blank image thanks to `cv2.draw_contours`. The function returns the array containing the contours, and the modified copies of the images.

```python
#Switch colums of the contour is needed for cv2.draw_contours to work properly.  
#This function simply switches columns 0 and 1.
def switch_col(ctr):
    for i in range (0,len(ctr)):
        switch_col = np.copy(ctr[i][:,0])
        ctr[i][:,0] = ctr[i][:,1]
        ctr[i][:,1] = switch_col
    return ctr

def draw_contours(indexes, c_images):
    for j in range (0, indexes.shape[1]):        
        ctr = [np.array(contours[0][int(indexes[0][j])]).reshape((-1,2)).astype(np.int32),
                   np.array(contours[1][int(indexes[1][j])]).reshape((-1,2)).astype(np.int32),
                   np.array(contours[2][int(indexes[2][j])]).reshape((-1,2)).astype(np.int32)]

        # Switch columns for cv2.drawContours
        ctr = switch_col(ctr)

        for i in range (0,len(ctr)):
            cv2.drawContours(c_images[i], [ctr[i]], -1, 255, -1)
    return ctr, c_images
```

## Counting the shapes and pixels of different colors
This function both counts the number of shapes of each color, and the number of pixels of each color. `list_centers` is a list with the coordinates of the centers of all the shapes. `blue_indexes`and `black_indexes` store the index (in `list_centers`) of blue and black shapes in each image. 

Detecting the color of the shape is done simply by looking at the blue channel of the original image, at the coordinates of the center of the shape. If the blue channel is above 100, the shape is blue. Else it is black. We noticed that the normalization was very well done, in a way that blue channels of black shapes are strictly 0. But we put a higher threshold for safety.

Then, the number of pixel of each color is counted after the function `draw_contours` is called. `draw_contours` returns images, in which a blue (or black) pixel has value 255, else 0. We just have to sum all pixels of those images, and divide by 255, to find the number of pixels corresponding to each color.

```python
blue_indexes = np.empty((0,5), int)
black_indexes = np.empty((0,5), int)

for i in range (0, len(list_centers)):
    blue_idx_temp = np.array([])
    black_idx_temp = np.array([])
    
    #detecting the color of the shape
    for j in range (0,len(list_centers[0])):
        if new_images[i, int(list_centers[i,j,0]), int(list_centers[i,j,1]), 2] > 100: #blue shape
            blue_idx_temp = np.append(blue_idx_temp,j)
            
        else : #black shape
            black_idx_temp = np.append(black_idx_temp,j)
    
    blue_indexes = np.vstack((blue_indexes, blue_idx_temp))
    black_indexes = np.vstack((black_indexes, black_idx_temp))
    #displaying the number of shapes of each colors for each image
    print("arena-shapes-0%i"%(i+1), ": Blue shapes :", blue_indexes.shape[1], ", Black shapes :", black_indexes.shape[1])
    
c_images_blue = [np.zeros_like(new_images[0]),np.zeros_like(new_images[1]),np.zeros_like(new_images[2])]
c_images_black = [np.zeros_like(new_images[0]),np.zeros_like(new_images[1]),np.zeros_like(new_images[2])]
    
#Draw the contours of all blue and black shapes
ctr_blue, c_images_blue = draw_contours(blue_indexes, c_images_blue)
ctr_black, c_images_black = draw_contours(black_indexes, c_images_black)

fig, ax = plt.subplots(3,3, figsize=(16,16))

#display images
for i in range (0,len(ctr_blue)):
    ax[i,0].imshow(c_images_blue[i], cmap='gray')
    ax[i,0].set_title('Blue shapes, arena-shapes-0%i)'%(i+1))

    ax[i,1].imshow(c_images_black[i], cmap='gray')
    ax[i,1].set_title('Black shapes, arena-shapes-0%i'%(i+1))

    ax[i,2].imshow(new_images[i], cmap='gray')
    ax[i,2].set_title('arena-shapes-0%i (normalized)'%(i+1))

#calculate the number of pixels
for i in range (0,len(ctr_blue)):
    print("Image %i :"%(i+1), "Blue pixels :", np.sum(c_images_blue[i]/255), "Black pixels", np.sum(c_images_black[i]/255))

plt.show()
```
